#Toy example with linear regression with 2 parameters

```python
from numpy import random

x = np.linspace(-20,20,100)
y = 7 * x + 13 + np.random.normal(0, 0.5, size = x.shape)
```

```python
def predict(w,b,data):
    return w * data + b
```

```python
def loss_funct(y,pred):
  return np.mean((y-pred)**2)
```

```python
l = loss_funct(y,predict(7,13,x))
print(l)
```

```python
alpha = 0.01
beta1 = 0.9
beta2 = 0.999
eps = 1e-8
w_init = np.random.randn()
b_init = np.random.randn()
```

```python
#wx + b
w = w_init
b = b_init
print(w,b)
w_m = 0.0
w_v = 0.0
b_m = 0.0
b_v = 0.0
n = len(x)
T = 2500

for t in range(1, T+1):
  pred = predict(w,b,x)
  error = pred - y
  loss = loss_funct(y,pred)

  dw = (2.0/n) * np.sum(error * x)
  db = (2.0/n) * np.sum(error)

  if (t%100 == 0):
    print("Epoch: " + str(t) + " Loss: " + str(loss))

  #update moments m
  w_m = beta1 * w_m + (1-beta1) * dw
  w_m_hat = w_m / (1-beta1**t)              #bias correction

  w_v = beta2 * w_v + (1-beta2) * dw**2
  w_v_hat = w_v / (1-beta2**t)

  #update moments b
  b_m = beta1 * b_m + (1-beta1) * db
  b_m_hat = b_m / (1-beta1**t)              #bias correction

  b_v = beta2 * b_v + (1-beta2) * db**2
  b_v_hat = b_v / (1-beta2**t)

  w -= alpha * w_m_hat / (np.sqrt(w_v_hat) + eps)
  b -= alpha * b_m_hat / (np.sqrt(b_v_hat) + eps)

print(w,b)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import cm
from IPython.display import HTML

# --------- Define loss and gradient ---------
def loss(w):
    return  w[0]**2 + 500 * w[1]**2

def grad(w):
    true_grad = np.array([2 * w[0], 1000 * w[1]])
    noise = np.random.normal(0, 10, size=2)
    return true_grad + noise


def grad_sparse(w,step):
  g = grad(w)
  if step % 5 != 0:
    g[0] = 0.0
  return g


# --------- Hyperparameters ---------
T = 120 # frames

w_init = np.array([5.0, 5.0])

# --------- Run optimizers and store trajectories ---------
sgd_path = []
adam_path = []
adag_path = []
rmsprop_path = []
momentum_path = []

#------------------------#
eps = 1e-8
alpha1 = 0.0005
# SGD
w = w_init.copy()
w = w * np.array([-1,1])
for t in range(T):
    sgd_path.append(np.append(w.copy(),loss(w)))
    g = grad_sparse(w,t)
    w -= alpha1 * g

#------------------------#
eps = 1e-8
alpha = 0.2
beta1 = 0.9
beta2 = 0.999
# Adam
w = w_init.copy()
m = np.zeros(2)
v = np.zeros(2)
for t in range(1, T + 1):
    adam_path.append(np.append(w.copy(),loss(w)))
    g = grad_sparse(w,t-1)
    m = beta1 * m + (1 - beta1) * g
    v = beta2 * v + (1 - beta2) * (g ** 2)
    m_hat = m / (1 - beta1**t)
    v_hat = v / (1 - beta2**t)
    w -= alpha * m_hat / (np.sqrt(v_hat) + eps)

#------------------------#
eps = 1e-8
eta = 0.5
# AdaGrad
g_accum = np.zeros_like(w)
w = w_init.copy()

for t in range(1, T+1):
    adag_path.append(np.append(w.copy(),loss(w)))
    g = grad_sparse(w,t-1)
    g_accum += g**2
    w -= eta * g / (np.sqrt(g_accum) + eps)

#------------------------#
eps = 1e-8
beta = 0.9
alpha = 0.1
#RMSProp
w = w_init.copy()
v = np.zeros(2)

for t in range(1, T + 1):
    rmsprop_path.append(np.append(w.copy(), loss(w)))
    g = grad_sparse(w,t-1)
    v = beta * v + (1 - beta) * (g ** 2)
    w -= alpha * g / (np.sqrt(v) + eps)

#------------------------#
beta1 = 0.9
alpha = 0.005
#Momentum
w = w_init.copy()
w = w * np.array([-1,1])
v = np.zeros(2)

for t in range(T):
    momentum_path.append(np.append(w.copy(), loss(w)))
    g = grad_sparse(w,t)
    v = beta1 * v + (1-beta1)*g
    w -= alpha * v


momentum_path = np.array(momentum_path)
rmsprop_path = np.array(rmsprop_path)
sgd_path = np.array(sgd_path)
adam_path = np.array(adam_path)
adag_path = np.array(adag_path)


```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import cm
from IPython.display import HTML

# --------- Define loss and gradient ---------
def loss(w):
    return w[0]**2 + 500 * w[1]**2

def grad(w):
    true_grad = np.array([2 * w[0], 1000 * w[1]])
    noise = np.random.normal(0, 20, size=2)
    return true_grad + noise


# --------- Hyperparameters ---------
alpha = 0.2
alpha1 = 0.0005
beta1 = 0.9
beta2 = 0.999
eps = 1e-8
T = 250 # frames
w_init = np.array([5.0, 5.0])

# --------- Run optimizers and store trajectories ---------

adam_with_bias_corr_path = []

# Adam
w = w_init.copy()
m = np.zeros(2)
v = np.zeros(2)
for t in range(1, T + 1):
    g = grad(w)
    m = beta1 * m + (1 - beta1) * g
    v = beta2 * v + (1 - beta2) * (g ** 2)
    m_hat = m / (1 - beta1**t)
    v_hat = v / (1 - beta2**t)
    w -= alpha * m_hat / (np.sqrt(v_hat) + eps)
    adam_with_bias_corr_path.append(np.linalg.norm(alpha * m_hat / (np.sqrt(v_hat) + eps)))


adam_without_bias_path = []

# Adam without bias corr
w = w_init.copy()
m = np.zeros(2)
v = np.zeros(2)
for t in range(1, T + 1):
    g = grad(w)
    m = beta1 * m + (1 - beta1) * g
    v = beta2 * v + (1 - beta2) * (g ** 2)
    w -= alpha * m / (np.sqrt(v) + eps)
    adam_without_bias_path.append(np.linalg.norm(alpha * m / (np.sqrt(v) + eps)))


import matplotlib.pyplot as plt



epochs = list(range(1, len(adam_with_bias_corr_path) + 1))

plt.plot(epochs, adam_with_bias_corr_path, label="Adam With Bias Correction", color='blue')
plt.plot(epochs, adam_without_bias_path, label="Adam Without Bias Correction ", color='green')

plt.xlabel("Epoch")
plt.ylabel("Step Size (L2 Norm)")
plt.title("Step Size Comparison: Adam vs Adam without Bias Correction")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
